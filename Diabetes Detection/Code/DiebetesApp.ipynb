```python
import pandas as pd
df= pd.read_csv("FinalDiabetesDataset.csv")
df.head(5)
```

```python
x = df.drop(['Outcome'],axis=1)
y = df['Outcome']
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
dt_model.fit(x,y)
print("Training Score : ",dt_model.score(x,y))
```

```python
import joblib
joblib.dump(dt_model,"DT.pkl")
dt = joblib.load("DT.pkl")
```
